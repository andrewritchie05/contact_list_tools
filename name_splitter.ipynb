```python
import pandas as pd
import numpy as np
import os
```

```python
# remove leading and trailing spaces

def space_stripper(dataframe):
    col_list = dataframe.select_dtypes(include='object').columns.tolist()
    
    for col in col_list:
        if len(dataframe[col]) > 0:
            col_array = np.array(dataframe[col].astype(str))
            stripped_col = np.array([])

            for idx in range(0,len(col_array)):
                item = col_array[idx]
                item.lstrip().rstrip()

                stripped_col = np.append(stripped_col, item)

            dataframe[col] = stripped_col
        
        else:
               return
            
        dataframe = dataframe.replace('nan', np.nan)
```

```python
#Split name string into first (fname), first middle (ffname), and last (lname)

def name_split(dataframe):
    names = np.array(dataframe['name'].astype(str))
    cols_list = dataframe.columns.tolist()


    ffname = np.array([])
    fname = np.array([])
    lname = np.array([])

    i = 0 
    while i < len(names):
         
        # remove '.' to ensure consistent formatting    
        names[i] = names[i].replace('.', '')
        
        name = names[i].split(' ')
        
        if len(name) == 1:
            f = name[0]
            ff = f
            l = ''
            
        elif len(name) == 2:
            f = name[0]
            ff = f
            l = name[1]
            
            try:
                # change SURNAME to Surname
                l = l[0]+l[1:].lower()
            except:
                l = l
            
        elif len(name) == 3:
            f = name[0]
            ff = name[0]+ ' ' + name[1]
            l = name[2]
            
            try:
                # change SURNAME to Surname
                l = l[0]+l[1:].lower()
            except:
                l = l

        ffname = np.append(ffname, ff)
        fname = np.append(fname, f)
        lname = np.append(lname, l)

        i += 1

    # insert the new columns at desired index
    dataframe.insert(1, 'ffname', ffname)   
    dataframe.insert(2, 'fname', fname) 
    dataframe.insert(3, 'lname', lname) 
    
    dataframe = dataframe.replace('nan', np.nan)
```

```python
# remember to edit dfs and paths

df_list = ['df1', 'df2', 'df3', 'df4']

for dataframe in df_list:
    
    df = pd.read_csv('...path/{0}.csv'.format(dataframe))
    
    space_stripper(df)
    name_split(df)

    df.to_csv('...path/cleaned/{0}.csv'.format(dataframe),index=False)
```
